```python
import pandas as pd
import datetime
import matplotlib.pyplot as plt
```

```python
zoster=pd.read_csv('zoster.csv')
zoster.columns=['PatientID','age','cen','date_string']
zoster['date']=pd.to_datetime(zoster.date_string)
```

```python
zoster.head()
```

```python
counts=zoster[['PatientID','date']].groupby(pd.Grouper(key='date', freq='1M')).count()

```

```python
plt.plot(counts.index,counts.PatientID)
plt.title('zoster diagnosis per month')
plt.show()
```

```python
months=['MAR','APR','MAY','JUN','JUL','AUG','SEP','OCT','NOV','DEC','JAN','FEB']
last_year=counts[-12:]
before_year=counts[-24:-12]
```

```python
plt.plot(months,last_year.PatientID)
plt.plot(months,before_year.PatientID)
plt.legend(labels=['2020-2021','2019-2020'])
plt.title('zoster diagnosis per month')
plt.show()
```
